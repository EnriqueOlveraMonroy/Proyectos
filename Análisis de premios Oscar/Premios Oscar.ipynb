## CARGA Y LIMPIEZA DEL DATA SET

```python
#Carga de Librer√≠a Pandas para limpieza y carga del Data Set

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/the_oscar_award 1927 -2023.csv",delimiter = ",", on_bad_lines='skip')
df.head(5)
```

```python
#Descripci√≥n general de los tipos de datos que contiene cada variable

df.info()
```

```python
#Revisamos valores nulos

df.isnull().sum()
```

```python
#Revisamos valores ausentes

df.isna().sum()
```

```python
#Eliminamos valoes ausentes y confirmamos que est√© completo el Data Set

df = df.dropna()
df.isna().sum()
```

```python
#Visualizamos el Data Set , est√° ordenado por a√±o de manera ascendente.

df.head(10)
```

```python
#Visualizaci√≥n de los √∫ltimos datos en el Data Set

df.tail(10)
```

```python
#Estandarizamos la columna "category" para que s√≥lo contenga MAY√öSUCLAS c√≥mo los primeros registros.
df["category"] = df["category"].str.upper()
df["category"]
```

```python
#Datos Generales

df.describe()
```

##ALGUNOS DATOS INTERESANTES

###Pel√≠culas ganadoras de premios Oscar

```python
Pel√≠culas_Ganadoras = df[df["winner"] == True]
Pel√≠culas_Ganadoras
```

###Pel√≠culas que han obtenido m√°s premios Oscar

```python
Top10_Ganadoras = Pel√≠culas_Ganadoras[["film","winner","year_film"]].groupby(["film","year_film"]).count()
Top10_Ganadoras = Top10_Ganadoras.sort_values("winner", ascending=False )
Top10_Ganadoras.head(10)
```

```python
#Ajuste de indice y columnas
Top10_Ganadoras = pd.DataFrame(Top10_Ganadoras.head(10))
Top10_Ganadoras = Top10_Ganadoras.reset_index()
Top10_Ganadoras.columns
```

```python
#Gr√°fico

sns.set_style("dark")
sns.set_context("talk")
plt.figure(figsize=(10, 6))

ax = plt.subplot()

sns.barplot(x='film', y='winner', data=Top10_Ganadoras, palette = ["black","red"])

#Poner t√≠tulo a mi gr√°fico
ax.set_title('PEL√çCULAS QUE HAN OBTENIDO M√ÅS PREMIOS OSCAR EN LA HISTORIA',fontsize=18, fontweight='bold')


#Nombrar y modificar eje X y eje Y
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('PEL√çCULAS GANADORAS', fontsize=14)
ax.set_ylabel('N√öMERO DE PREMIOS OSCAR GANADOS', fontsize=14)

#Agregar valores a las barras
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=12)

#Guardar mi gr√°fico
plt.savefig('gr√°fico1.png')

#Mostrar gr√°fico
plt.show()
```

###Pel√≠culas con mayor n√∫mero de nominaciones en los premios Oscar




```python
Top10_Nominaciones = df[["film","winner","year_film"]].groupby(["film","year_film"]).count()
Top10_Nominaciones = Top10_Nominaciones.sort_values("winner", ascending=False )
Top10_Nominaciones.head(10)
```

```python
#Ajuste de indice y columnas
Top10_Nominaciones = pd.DataFrame(Top10_Nominaciones.head(10))
Top10_Nominaciones = Top10_Nominaciones.reset_index()
Top10_Nominaciones.columns
```

```python
#Gr√°fico

sns.set_style("dark")
sns.set_context("talk")
plt.figure(figsize=(10, 6))

ax = plt.subplot()

sns.barplot(x='film', y='winner', data=Top10_Nominaciones, palette = ["gray","red"])

#Poner t√≠tulo a mi gr√°fico
ax.set_title('PEL√çCULAS QUE HAN OBTENIDO M√ÅS NOMINACIONES A LOS PREMIOS OSCAR EN LA HISTORIA',fontsize=18, fontweight='bold')


#Nombrar y modificar eje X y eje Y
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('PEL√çCULAS NOMINADAS', fontsize=14)
ax.set_ylabel('N√öMERO DE NOMINACIONES', fontsize=14)

#Agregar valores a las barras
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=12)

#Guardar mi gr√°fico
plt.savefig('gr√°fico2.png')

#Mostrar gr√°fico
plt.show()
```

###Pel√≠culas con m√°s nominaciones y los premios ganados

```python
Lista_Nominadas = (list(Top10_Nominaciones["film"]))
N√∫mero_Ganadas = Pel√≠culas_Ganadoras[Pel√≠culas_Ganadoras["film"].isin(Lista_Nominadas)]
N√∫mero_Ganadas = N√∫mero_Ganadas[["film","winner"]].groupby("film").count()
N√∫mero_Ganadas = N√∫mero_Ganadas.sort_values("film", ascending = True)
N√∫mero_Ganadas = N√∫mero_Ganadas.reset_index()
N√∫mero_Ganadas
```

```python
Top10_Nom_Ganadas = Top10_Nominaciones.rename(columns = {'winner' : 'nomination'}, inplace = True)
Top10_Nom_Ganadas = Top10_Nominaciones.sort_values("film")
Top10_Nom_Ganadas = Top10_Nom_Ganadas.reset_index()
Top10_Nom_Ganadas
```

```python
Top10_Nom_Ganadas["Real_winners"] = N√∫mero_Ganadas["winner"]
Top10_Nom_Ganadas
```

```python
#Gr√°fico

sns.set_style("dark")
sns.set_context("talk")
plt.figure(figsize=(12, 6))

ax = plt.subplot()

sns.barplot(x='film', y='nomination', data=Top10_Nom_Ganadas, color = "red")
sns.barplot(x='film', y='Real_winners', data=Top10_Nom_Ganadas, color ="black")

#Poner t√≠tulo a mi gr√°fico
ax.set_title('PEL√çCULAS QUE HAN OBTENIDO M√ÅS NOMINACIONES Y SUS PREMIOS GANADOS',fontsize=18, fontweight='bold')


#Nombrar y modificar eje X y eje Y
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('PEL√çCULAS', fontsize=14)
ax.set_ylabel('N√öMERO DE NNOMINACIONES/OSCAR', fontsize=14)

#Agregar valores a las barras
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=12)



#Guardar mi gr√°fico
plt.savefig('gr√°ficoNom_Ganadas.png')

#Mostrar gr√°fico
plt.show()
```

###Pel√≠culas que ganaron todos los premios a los que fueron nominadas


```python
Premios = Pel√≠culas_Ganadoras[["film","winner","year_film"]].groupby(["film","year_film"]).count()
Nominaciones = df[["film","winner","year_film"]].groupby(["film","year_film"]).count()

Premios["N√∫mero_Nominaciones"] = Nominaciones["winner"]


Top10_Ganadoras_Nominadas = Premios[Premios["winner"] == Premios["N√∫mero_Nominaciones"]].sort_values("winner", ascending = False)
Top10_Ganadoras_Nominadas = Top10_Ganadoras_Nominadas.head(10)
Top10_Ganadoras_Nominadas
```

```python
#Ajuste de indice y columnas
Top10_Ganadoras_Nominadas = Top10_Ganadoras.reset_index()
Top10_Ganadoras_Nominadas.columns
```

```python
#Gr√°fico

sns.set_style("dark")
sns.set_context("talk")
plt.figure(figsize=(10, 6))

ax = plt.subplot()

sns.barplot(x='film', y='winner', data=Top10_Ganadoras_Nominadas, color = "red")

#Poner t√≠tulo a mi gr√°fico
ax.set_title('PEL√çCULAS QUE HAN GANADO TODOS LOS PREMIOS A LAS QUE FUERON NOMINADAS',fontsize=18, fontweight='bold')


#Nombrar y modificar eje X y eje Y
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('PEL√çCULAS NOMINADAS/GANADORAS', fontsize=14)
ax.set_ylabel('N√öMERO DE PREMIOS OSCAR GANADOS', fontsize=14)

#Agregar valores a las barras
for i in ax.containers:
    ax.bar_label(i, label_type='edge', fontsize=12)

#Guardar mi gr√°fico
plt.savefig('gr√°fico3.png')

#Mostrar gr√°fico
plt.show()
```

###Nominaciones para los premios Oscar 2023


```python
Nominaciones_2023 = df[["year_film","film","category", "name","winner"]]
Nominaciones_2023 = Nominaciones_2023[Nominaciones_2023["year_film"] == 2022]
Nominaciones_2023
```

###Ganadores de premios Oscar 2023

```python
Ganadores_2023 = Nominaciones_2023[Nominaciones_2023["winner"]==True]
Ganadores_2023
```

###Top de ganadores 2023

```python
Top_Ganadores2023 = Ganadores_2023[["film","winner"]].groupby("film").count()
Top_Ganadores2023 = Top_Ganadores2023.sort_values("winner",ascending = False)
Top_Ganadores2023
```

```python
#Ajuste de indice y columnas
Top_Ganadores2023 = Top_Ganadores2023.reset_index()
Top_Ganadores2023.columns
```

```python
categorias = list(Top_Ganadores2023["film"])
valores = list(Top_Ganadores2023["winner"])

#Set para el doughnut
fig, ax = plt.subplots()
fig.set_size_inches(8, 8)

#Gr√°fico
ax.pie(valores, labels=categorias, colors=sns.color_palette('dark'), wedgeprops={'width': 0.5})
circulo_central = plt.Circle((0,0), 0.3, color='white')
ax.add_artist(circulo_central)

#T√≠tulo
ax.set_title('GANADORES DE PREMIOS OSCAR 2023')
ax.axis('equal')

#Guardar mi gr√°fico
plt.savefig('gr√°fico4.png')

plt.show()
```

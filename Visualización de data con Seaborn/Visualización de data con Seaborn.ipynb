## 2.1. Visualización de variables cuantitativas


```python
# CREE UNA COLUMNA ADICIONAL CON EL NOMBRE DEL DÍA DE CADA UNA DE LAS FECHAS PRESENTES

dataTotal["diasdelasemana"] = dataTotal["fecha"].apply(lambda x: x.dayofweek)
```

```python
#SE LE ASIGNA EL NOMBRE

mapeo = {0 : "Lunes", 1: "Martes", 2:"Miércoles", 3:"Jueves", 4: "Viernes", 5: "Sabado", 6: "Domingo"}
```

```python
dataTotal["diasdelasemana"] = dataTotal["diasdelasemana"].map(mapeo)
```

```python
# LOS DÍAS DE LA SEMANA QUE MENOS O MÁS TIENEN AFLUENCIA EN TOTAL

sns.relplot(data=dataTotal, x="mes", y="TOTALES", hue="diasdelasemana")
plt.show()
```

```python
# LOS DÍAS DE LA SEMANA QUE MENOS O MÁS TIENEN AFLUENCIA EN TREN LIGERO

sns.relplot(data=dataTotal, x="mes", y="trenligero_afluencia_total", hue="diasdelasemana")

plt.show()
```

```python
# LOS DÍAS DE LA SEMANA QUE MENOS O MÁS TIENEN AFLUENCIA EN CABLEBUS

sns.relplot(data=dataTotal, x="mes", y="cablebus_afluencia_total", hue="diasdelasemana")

plt.show()
```

```python
# LOS DÍAS DE LA SEMANA QUE MENOS O MÁS TIENEN AFLUENCIA EN TROLEBUS

sns.relplot(data=dataTotal, x="mes", y="trolebus_afluencia_total", hue="diasdelasemana")

plt.show()
```

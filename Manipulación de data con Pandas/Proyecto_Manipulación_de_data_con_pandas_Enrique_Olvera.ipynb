### Artistas más populares

Tienes muchos artistas únicos en este género, primero selecciona los 5

```python
artistas_mas_populares = tracks_validos[['artista_principal', 'popularidad']]\
                         .groupby('artista_principal')\
                         .max().sort_values('popularidad', ascending=False)\
                         .head(5).reset_index()
artistas_mas_populares
```

```python
artistas_plot = list(artistas_mas_populares['artista_principal'])
artistas_plot
```

```python
tracks_plot = tracks_validos.loc[tracks_validos['artista_principal'].isin(artistas_plot)]
tracks_plot = tracks_plot.loc[tracks_plot['año']>'2010']
tracks_plot.head()
```

```python
tracks_plot.shape
```

```python
tracks_artista = tracks_plot[['artista_principal', 'popularidad']].groupby('artista_principal').mean()

tracks_artista
```

```python
tracks_agrupado = tracks_plot[['artista_principal',
                               'año',
                               'popularidad']]\
    .groupby(['año', 'artista_principal'])\
    .agg(
    {
        'popularidad': 'mean',
    }
)
tracks_agrupado
```

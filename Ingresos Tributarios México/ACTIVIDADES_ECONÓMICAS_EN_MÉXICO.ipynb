```python
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick


import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/DATA_SCIENCE_SP/IMPUESTOS/ACT_ECONOMICAS/SectorActEco.csv", encoding='ISO-8859-1')
df.head()
```

```python
df['Año'] = pd.to_datetime(df['Año'].astype("str"), format = "%Y%")
df
```

```python
lista_columnas = df.columns

for indice, valor in enumerate(lista_columnas):
    print(indice, valor)
```

En seaborn, "le7" significa que la escala del eje está en notación científica, con un factor de multiplicación de 10 elevado a la 7 (es decir, decena de millón). Por ejemplo, si tiene datos que varían entre 1 y 10,000,000, seaborn podría mostrar la escala del eje como "1e+07", lo que significa que los valores se multiplican por diez millones para mostrarlos en el gráfico.

La escala "le7" se utiliza a menudo para visualizar datos que tienen una gran variación entre los valores, como los datos financieros o los datos de tamaño de archivo. La notación científica permite mostrar valores grandes en una escala manejable y puede hacer que los gráficos sean más fáciles de interpretar.

```python
#Variables a graficar
Xbar = lista_columnas[1:]
Ybar = df.iloc[-1:,1:].values.flatten()

#Preparar el plot
sns.set_style("dark")
sns.set_context("talk")
plt.figure(figsize=(25, 6))


#Graficar
sns.barplot(x=Xbar, y=Ybar)

ax = plt.subplot()


#Poner título a mi gráfico
ax.set_title('Número de Contribuyentes Activos',fontsize=18, fontweight='bold')


#Nombrar y modificar eje X y eje Y
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Actividades Financieras', fontsize=14)
ax.set_ylabel('Contribuyentes Activos en 2022', fontsize=14)

#Cambiar la escala del eje Y
ax.ticklabel_format(style='plain', axis='y')

#Formato
fmt = '{x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.show()
```

```python
# Crea una figura con 2 filas y 3 columnas de subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(30, 12), dpi=100)


# Crea datos de ejemplo para cada gráfico
x = df.iloc[:,0]
y1 = df[lista_columnas[1]]
y2 = df[lista_columnas[2]]
y3 = df[lista_columnas[3]]
y4 = df[lista_columnas[4]]
y5 = df[lista_columnas[5]]
y6 = df[lista_columnas[6]]


# Grafica cada conjunto de datos en un subplot separado
axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[0, 2].plot(x, y3)
axs[1, 0].plot(x, y4)
axs[1, 1].plot(x, y5)
axs[1, 2].plot(x, y6)


# Agrega etiquetas a los subplots y a la figura
fig.suptitle('Actividades Económicas con mayor número de usuarios activos')
axs[0, 0].set(title= str(y1.name), xlabel='Año', ylabel='Contribuyentes Activos')
axs[0, 1].set(title= str(y2.name), xlabel='Año', ylabel='Contribuyentes Activos')
axs[0, 2].set(title= str(y3.name), xlabel='Año', ylabel='Contribuyentes Activos')
axs[1, 0].set(title= str(y4.name), xlabel='Año', ylabel='Contribuyentes Activos')
axs[1, 1].set(title= str(y5.name), xlabel='Año', ylabel='Contribuyentes Activos')
axs[1, 2].set(title= str(y6.name), xlabel='Año', ylabel='Contribuyentes Activos')


#Truco para ajustar las gráficas y que no se estorben entre ellas
plt.tight_layout()


plt.show()
```

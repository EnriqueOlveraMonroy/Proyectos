#Recaudación | Ingresos tributarios del Gobierno Federal

Ingresos por impuesto (Millones de pesos)

Notas: Otros ingresos tributarios: incluye el Impuesto Empresarial a Tasa Única (IETU), Impuesto a los Depósitos en Efectivo (IDE), Impuesto Sobre Tenencia o Uso de Vehículos (ISTUV), Accesorios y a partir de 2019 el Impuesto al Activo (IMPAC).
Impuesto Sobre la Renta (ISR): A partir de 2015 incluye el ISR de contratistas y asignatarios. De 2010 a 2018 incluye el Impuesto al Activo (IMPAC).

```python
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np

import matplotlib.ticker as mtick
```

```python
#CARGAMOS DATS SET
df = pd.read_csv("/content/IngresosTributarios.csv", encoding='ISO-8859-1')
df
```

```python
df.replace({"n.d.": 0}, inplace=True)
df
```

#Comportamiento de la recaudación de impuestos totales a través de los años 2010 - 2023

```python
df_total_anual = df.groupby("Año ")["Total de Ingresos tributarios"].sum()
df_total_anual
```

```python
plt.style.use('ggplot')
df_total_anual.plot(kind='bar', title='Recaudación de impuestos total acumulada desde 2010 - 2023', xlabel='Años', ylabel="Billones de Pesos", figsize=(10, 6))
plt.show()
```

#Comportamiento de la recaudación de impuesto ISR a través de los años 2010 - 2023

```python
df_anual_ISR = df.groupby("Año ")["Impuesto Sobre la Renta"].sum()
df_anual_ISR
```

```python
plt.style.use('ggplot')
df_anual_ISR.plot(kind='bar', title='Recaudación de ISR total acumulada desde 2010 - 2023', xlabel='Años', ylabel="Billones de Pesos", figsize=(10, 6))
plt.show()
```

#Comportamiento de la recaudación de impuesto IVA a través de los años 2010 - 2023

```python
df_anual_IVA = df.groupby("Año ")["Impuesto al Valor Agregado"].sum()
df_anual_IVA
```

```python
plt.style.use('ggplot')
df_anual_IVA.plot(kind='bar', title='Recaudación de IVA acumulada desde 2010 - 2023', xlabel='Años', ylabel="Billones de Pesos", figsize=(10, 6))
plt.show()
```

#Comportamiento de la recaudación de impuesto IEPS a través de los años 2010 - 2023

```python
df_anual_IEPS = df.groupby("Año ")["Impuesto Especial sobre Producción y Servicios"].sum()
df_anual_IEPS
```

```python
plt.style.use('ggplot')
df_anual_IEPS.plot(kind='bar', title='Recaudación de IEPS acumulada desde 2010 - 2023', xlabel='Años', ylabel="Millones de Pesos", figsize=(10, 6))
plt.show()
```

#¿Qué producto recuadó más IEPS en el año 2022?

```python
df_IEPS_2022 = df.groupby("Año ")[['Gasolinas y diésel',
       'Tabacos labrados', 'Alcohol', ' Cerveza y bebidas refrescantes',
       'Juegos y sorteos', 'Telecomunicaciones', 'Bebidas energetizantes',
       ' Bebidas saborizadas\r',
       'Alimentos no básicos con alta densidad calórica ', 'Plaguicidas\r',
       'Carbono\r']].sum()

df_IEPS_2022
```

```python
df_IEPS_2022 = df_IEPS_2022.iloc[-2:-1,:]
df_IEPS_2022
```

```python
plt.style.use('ggplot')
df_IEPS_2022.plot(kind='bar', title='Recaudación de IEPS acumulada 2022', xlabel='Año', ylabel="Millones de Pesos", figsize=(10, 6))
plt.show()
```

#¿Cuánto se recuadó en total de ISR, IVA y IEPS en el año 2022?




```python
df_impuestos_anuales2022 = df.groupby("Año ")[['Impuesto Sobre la Renta', 'Impuesto al Valor Agregado',
       'Impuesto Especial sobre Producción y Servicios']].sum()

df_impuestos_anuales2022 = df_impuestos_anuales2022.iloc[-2:-1,:]
df_impuestos_anuales2022
```

```python
plt.style.use('ggplot')
df_impuestos_anuales2022.plot(kind='bar', title='Recaudación de ISR, IVA y IEPS 2022', xlabel='Año', ylabel="Billones de Pesos", figsize=(10, 6))
plt.show()
```

#¿Cuánto se recuadó por mes de ISR, IVA y IEPS en el año 2022?


```python
df_impuestos_mensuales2022 = df[['Año ', 'Mes','Impuesto Sobre la Renta', 'Impuesto al Valor Agregado',
       'Impuesto Especial sobre Producción y Servicios']]

df_impuestos_mensuales2022 = df_impuestos_mensuales2022[df_impuestos_mensuales2022["Año "] == 2022]
df_impuestos_mensuales2022
```

```python
# Crea una figura con 2 filas y 3 columnas de subplots
plt.style.use('ggplot')

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(30, 12), dpi=100)


# Crea datos de ejemplo para cada gráfico
x = df_impuestos_mensuales2022["Mes"]
y1 = df_impuestos_mensuales2022['Impuesto Sobre la Renta']
y2 = df_impuestos_mensuales2022['Impuesto al Valor Agregado']
y3 = df_impuestos_mensuales2022['Impuesto Especial sobre Producción y Servicios']




# Grafica cada conjunto de datos en un subplot separado
axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, y3)



# Agrega etiquetas a los subplots y a la figura
fig.suptitle('RECAUDACIÓN DE ISR, IVA Y IEPS EN EL AÑO 2022')
axs[0].set(title= str(y1.name), xlabel='Meses del Año 2022', ylabel='Millones de Pesos')
axs[1].set(title= str(y2.name), xlabel='Meses del Año 2022', ylabel='Millones de Pesos')
axs[2].set(title= str(y3.name), xlabel='Meses del Año 2022', ylabel='Millones de Pesos')



#Truco para ajustar las gráficas y que no se estorben entre ellas
plt.tight_layout()

plt.show()
```
